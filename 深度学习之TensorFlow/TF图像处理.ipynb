```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import tensorflow as tf
```

```python
image_raw_data_jpg = tf.gfile.FastGFile('data/1.jpg', 'rb').read()
```

# 1. 图像翻转处理

```python
with tf.Session() as sess:
    img_data_jpg = tf.image.decode_jpeg(image_raw_data_jpg)
    img_data_jpg = tf.image.convert_image_dtype(img_data_jpg, dtype=tf.float32)
    img_0 = img_data_jpg
    img_1 = tf.image.flip_up_down(img_data_jpg)
    img_2 = tf.image.flip_left_right(img_data_jpg)
    img_3 = tf.image.transpose_image(img_data_jpg)
    
    plt.figure(0)
    plt.title('original image',fontsize='large')
    plt.imshow(img_0.eval())
    
    plt.figure(1)
    plt.title('flip_up_down',fontsize='large')
    plt.imshow(img_1.eval())

    plt.figure(2)
    plt.title('flip_left_right',fontsize='large')
    plt.imshow(img_2.eval())
    
    plt.figure(3)
    plt.title('transpose_image',fontsize='large')
    plt.imshow(img_3.eval())
    plt.show()
```

# 2. 图像亮度变化

```python
with tf.Session() as sess:
    img_data_jpg = tf.image.decode_jpeg(image_raw_data_jpg)
    img_data_jpg = tf.image.convert_image_dtype(img_data_jpg, dtype=tf.float32)
    img_0 = img_data_jpg
    img_1 = tf.image.adjust_brightness(img_data_jpg, -0.3)
    img_2 = tf.image.adjust_brightness(img_data_jpg, 0.3)
    img_3 = tf.image.adjust_brightness(img_data_jpg, 0.5)
    
    plt.figure(0)
    plt.title('original image',fontsize='large')
    plt.imshow(img_0.eval())
    
    plt.figure(1)
    plt.title('-0.3',fontsize='large')
    plt.imshow(img_1.eval())

    plt.figure(2)
    plt.title('+0.3',fontsize='large')
    plt.imshow(img_2.eval())
    
    plt.figure(3)
    plt.title('+0.5',fontsize='large')
    plt.imshow(img_3.eval())
    plt.show()
```
